```
def AverPerTran(cusID):
  Transaction=data1[data1['csn']==cusID]['transaction_info']
  numOfTran=len(Transaction)
  NumAr_Sum=0
  priceSum=0
  for i in range(numOfTran):
    ti=Transaction.iloc[i]
    str = ti.replace("\'", "\"")
    with open('f.json', 'w') as f:
      f.write(str)
    with open('f.json') as f:
      ti = json.load(f)
      numOfArPerTran=len(ti)
      NumAr_Sum=NumAr_Sum+numOfArPerTran
      for k in range(numOfArPerTran):
        priceSum=priceSum+ti[k]['price']
  numOfArPerTran_Aver=NumAr_Sum/numOfTran
  priceAverPerTran=priceSum/numOfTran
  return numOfArPerTran_Aver, priceAverPerTran
```

```
from tqdm import tqdm
PerTran=[]
for i in tqdm(Customer1):
  PerTran.append(AverPerTran(i))
PerTran=np.array(PerTran)
```

```
from datetime import datetime
```

```
def fre(cusID):
  for i in range(len(data1.csn)):
    if data1.iloc[i].csn==cusID:
      startTime=data1.date[i]
      break
  for i in range(len(data1.csn)):
    if data1.iloc[i].csn==cusID:
      endTime=data1.date[i]
  date_format = "%m/%d/%Y"
  a = datetime.strptime(startTime, date_format)
  b = datetime.strptime(endTime, date_format)
  dis = b - a
  if dis.days==0:
    fre=0
  else:
    fre=(dis.days)/cus1Fre[cusID]
  return fre
```
